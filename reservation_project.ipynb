```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('bookings.csv', sep=';')
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.isna().sum()
```

```python
# Пользователи из каких страны показали наибольшее количество удачных бронирований?

df.query('is_canceled == 0').country.value_counts()[:10]
```

```python
# Сколько ночей в среднем бронируют в разбивке на тип отеля?

df.groupby('hotel', as_index = False).agg({'stays_total_nights':'mean'}).round(2)
```

```python
df.arrival_date_year.unique()
```

```python
df.head()
```

```python
# На какой месяц чаще всего успешно оформляли бронь в разбивке на каждый год?

df.query('is_canceled == 0').groupby(['arrival_date_year', 'arrival_date_month']).size().sort_values(ascending=False)[:3]
```

```python
# На какой месяц бронирования отеля типа Resort Hotel отменялись чаще всего в 2015, 2016, 2017?

df.query('is_canceled == 1 and hotel == "Resort Hotel"').groupby('arrival_date_year').size().sort_values(ascending=False)[:1]
```

```python
# Какой процент гостей отеля приезжает с детьми?

df['kids'] = df['children'] + df['babies']
(((df['kids'] > 0).sum() / len(df)) * 100).round(2)
```
